```python
inbound_path="abfss://devcontainer@prasannastgacctb73.dfs.core.windows.net/inbound/"
outbound_path="abfss://devcontainer@prasannastgacctb73.dfs.core.windows.net/outbound/"
dbutils.fs.ls(inbound_path)
dbutils.fs.ls(outbound_path)
```

```python
# here my requirement is i want to copy only csv files from the source to the destination

#for that first i need to get files list from source

files_list=dbutils.fs.ls(inbound_path)

# now i want to copy only csv files from the list and one by one copy to the destination
for file in files_list:
    if file.name.endswith("csv"):
        dbutils.fs.cp(file.path,outbound_path)
```
